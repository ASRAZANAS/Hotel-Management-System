```
# نظام إدارة الغرف في الفندق
hotel_rooms = []

def add_room():
    room_number = input("Enter Room Number: ")
    room_type = input("Enter Room Type (Single/Double/Suite): ")
    price = input("Enter Price per Night: ")

    # استخدام Dictionary لتخزين بيانات الغرفة
    room = {
        "number": room_number,
        "type": room_type,
        "price": price
    }
    hotel_rooms.append(room)
    print("Room added successfully!")

def display_rooms():
    if not hotel_rooms:
        print("\nNo rooms found in the system.")
    else:
        print("\n--- Hotel Room List ---")
        for r in hotel_rooms:
            print(f"Room: {r['number']} | Type: {r['type']} | Price: {r['price']} SR")

def search_room():
    search_num = input("Enter Room Number to search: ")
    found = False
    for r in hotel_rooms:
        if r['number'] == search_num:
            print(f"Found! Type: {r['type']}, Price: {r['price']} SR")
            found = True
            break
    if not found:
        print("Room not found.")

def main():
    while True:
        print("\n--- Hotel Management System ---")
        print("1. Add New Room")
        print("2. Display All Rooms")
        print("3. Search for a Room")
        print("4. Exit")

        choice = input("Select an option (1-4): ")

        # استخدام Conditional Statements
        if choice == '1':
            add_student() # هنا سيتم استدعاء دالة الإضافة
        elif choice == '2':
            display_rooms()
        elif choice == '3':
            search_room()
        elif choice == '4':
            print("Exiting System... Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

# تشغيل البرنامج
if __name__ == "__main__":
    main()
```
